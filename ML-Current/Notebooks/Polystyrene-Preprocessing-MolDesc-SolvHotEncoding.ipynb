```python
#============================================================================
# Helper function to parse multi-value text columns
# (for Solvents, we keep up to 'max_splits' in order)
#============================================================================
def parse_multivalue_text(cell_value, max_splits=4):
    """
    Parse a comma-separated list of text (e.g., solvents).
    If fewer entries than max_splits, fill the rest with None.
    If more, truncate.
    """
    if pd.isnull(cell_value) or str(cell_value).strip() == "":
        return [None] * max_splits
    
    parts = [p.strip() for p in str(cell_value).split(";")]
    if len(parts) < max_splits:
        parts += [None] * (max_splits - len(parts))
    else:
        parts = parts[:max_splits]
    return parts
```

```python
#============================================================================
# Detector handling - 
# I want to retain as much experimental data as possible so Detector column was converted from strings to a count of the detectors.
# Currently used outright in model - but thought is maybe I can incorporate it into quality score or the statistical side of the ML
#============================================================================
if "Detector" in data_clean.columns:
    data_clean["Detector"] = data_clean["Detector"].fillna("None")
    data_clean["Number_of_Detectors"] = data_clean["Detector"].apply(
        lambda x: 0 if x.strip().lower() == "none" else len(x.split(","))
    )

data_clean.drop("Detector", axis=1, inplace=True)
```

```python
#============================================================================
# Helper function to parse multi-value numeric columns
# Used with columns containing multiple numeric entries
# eg. Pore size / particle diameter / ratios
#============================================================================
def parse_multivalue_numeric(cell_value, max_splits=4):
    """
    Parse a comma-separated list of numeric values.
    - If there's only 1 value, the rest become NaN
    - If more than max_splits, truncate
    """
    if pd.isnull(cell_value) or str(cell_value).strip() == "":
        return [np.nan] * max_splits
    
    parts = str(cell_value).split(",")
    floats = []
    for part in parts:
        part = part.strip()
        if part == "":
            floats.append(np.nan)
        else:
            try:
                floats.append(float(part))
            except ValueError:
                floats.append(np.nan)
    
    if len(floats) < max_splits:
        floats += [np.nan] * (max_splits - len(floats))
    else:
        floats = floats[:max_splits]
    
    return floats
```

```python
# Stack them into ONE Series (ignore_index=True to re-label row index)
X_combined_series = pd.concat(
    [imputated['Solvents_0'], imputated['Solvents_1']],
    axis=0,
    ignore_index=True
)

# Convert that Series into a single-column DataFrame
X_combined = X_combined_series.to_frame(name='Solvents')

# Fit the OneHotEncoder on that one-column DataFrame
encoder = OneHotEncoder(
    categories=[UniqueSolvents_list], 
    handle_unknown='ignore',
    sparse_output=True
)
encoder.fit(X_combined)
X1_df = imputated[['Solvents_0']].rename(columns={'Solvents_0': 'Solvents'})
X2_df = imputated[['Solvents_1']].rename(columns={'Solvents_1': 'Solvents'})

#  Now transform each column separately (which is 1 feature)
# This is the hot encoding of the first solvent in the binary pair
X1 = encoder.transform(X1_df)

# This is the hot encoding of the second solvent in the binary pair
X2 = encoder.transform(X2_df)
```

```python
# This is the hot encoding of the first solvent in the binary pair
# 172 rows with 20 columns - only 20 unique solvents in polystyrene subset
X1
```

```python
# This is pulling the solvent 1 ratio and making it 1 dimensional.
ratio_1 = imputated['Solvent Ratio_1'].values.reshape(-1, 1)

# This is pulling the solvent 2 ratio and making it 1 dimensional.
ratio_2 = imputated['Solvent Ratio_2'].values.reshape(-1, 1)
```

```python
# This is taking the 1 / 0 hot encoding and turning it to the solvent hot encoding value being the ratio of that solvent.
# The solvent 1 hot encoded 1 dimensional array is multiplied by the solvent 1 ratio 1 dimensional array
X1_ratio = X1.multiply(ratio_1)

# The solvent 2 hot encoded 1 dimensional array is multiplied by the solvent 2 ratio 1 dimensional array
X2_ratio = X2.multiply(ratio_2)
```

```python
# This addition turns the two individual solvent hot encoding of their ratios into one array containing the hot encoding of the binary solvent pairs by their respective ratios.
X = X1_ratio + X2_ratio
X
```

```python
# This converts the strange compressed matrix of dtype into an array that pandas can use.
X_dense = X.toarray()

# Encoded df merges the array of hot encoded solvents by their ratios with the encoder feature names - not lovely with the Solvents_SolventName styling
encoded_df = pd.DataFrame(X_dense, columns=encoder.get_feature_names_out())
encoded_df
```

```python
# Replaces the encoder feature names headers with Unique Solvents list created earlier - allowing names to be cleaner.
encoded_df.columns = UniqueSolvents_list
encoded_df
```

```python
# Index reset in preparation for concatenation with final containing the reset of the seed data and descriptors
encoded_df = encoded_df.reset_index(drop=True)
imputated = imputated.reset_index(drop=True)

# final_encoded contains hot encoded solvents by their ratios with the rest of the preprocessed seed data containing descriptors
final_encoded = pd.concat([encoded_df, imputated], axis=1)
```

```python
# final_encoded contains hot encoded solvents by their ratios with the rest of the preprocessed seed data containing descriptors
# Dropping the solvent 0 and solvent 1 names columns since they've been hot encoded.
final_encoded = final_encoded.drop(columns=['Solvents_0', 'Solvents_1', 'Solvent Ratio_1', 'Solvent Ratio_2'])

final_encoded
```

```python
#============================================================================
# One-hot encode other single-valued categorical columns
#============================================================================
other_categorical_cols = [
    #"Polymer",
    "Stationary Phase",
    "Base Material",
    "Base Material Modification",
    "Phase",
    
]

other_categorical_cols = [c for c in other_categorical_cols if c in final_encoded.columns]

# Fill missing with "None" for these columns too
for c in other_categorical_cols:
    final_encoded[c] = final_encoded[c].fillna("None")

ohe_other = OneHotEncoder(sparse_output=False, drop='first')
cat_df = final_encoded[other_categorical_cols].astype(str)
encoded_array_other = ohe_other.fit_transform(cat_df)

encoded_feature_names_other = ohe_other.get_feature_names_out(other_categorical_cols)
encoded_other_df = pd.DataFrame(encoded_array_other, columns=encoded_feature_names_other)

# Drop original categorical columns
final_encoded.drop(columns=other_categorical_cols, inplace=True)

# Combine categorical column encoding to the final_encoded containing solvent encoding by ratio and seed data preprocessing
final_encoded = pd.concat([final_encoded.reset_index(drop=True), encoded_other_df], axis=1)
```
